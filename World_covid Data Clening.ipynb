## Numerical missing

```python
num_var = df.select_dtypes(include=['int64', 'float64']).columns
num_var
```

```python
df[num_var].isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean')
```

```python
imputer.fit(df[num_var])
```

```python
imputer.statistics_
```

```python
imputer.transform(df[num_var])
```

```python
df[num_var] = imputer.transform(df[num_var])
```

```python
df[num_var].isnull().sum()
```

## Categorical data

```python
cat_var = df.select_dtypes(include=['object']).columns
cat_var
```

```python
df[cat_var].isnull().sum()
```

```python
imputer = SimpleImputer(strategy = 'most_frequent')
imputer
```

```python
imputer.fit(df[cat_var])
```

```python
imputer.transform(df[cat_var])
```

```python
df[cat_var] = imputer.transform(df[cat_var])
```

```python
df[cat_var].isnull().sum()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df[cat_var].isnull())
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df[num_var].isnull())
```
